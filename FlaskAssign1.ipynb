```python
# Question 5
# url_for() function is used in Flask for url building.

from flask import Flask, redirect, url_for

flaskApp = Flask(__name__)

@flaskApp.route('/user/<Name>')
def UserHomePage(Name):
    return f"This is Home Page of User {Name}"

@flaskApp.route('/User/<Name>/<PageType>')
def UserPageType(PageType, Name):
    if PageType == 'ShopKart':
        return f"This is Shopping Kart List of user {Name}"
    elif PageType == 'Purchase':
        return f"This is purchased item list of user {Name}"
    
@flaskApp.route("/accountType/<Type>")
def userType(Type):
    if Type == 'premium':
        return redirect(url_for('UserHomePage', Name = Type))
    
    else:
        return redirect(url_for('UserPageType', Name = Type, PageType = 'Purchase'))
    
if __name__ == '__main__':
    flaskApp.run('0.0.0.0')
```
